# Individual zipcode power readings data preprocessing

Here we will undergo some basic data processing for each zipcode as follow:

## Timestamp alignment and PV components pivoting

Now, we look at the data with a specific zipcode. We need to read timestamp originally in UTC time, convert it as local time, and adjust data to align with 00, 15, 30, 45 minute marks. One can pivot timeseries data into a table where rows represent timestamps and columns represent PV component IDs. 

```python
#specify data source 
datasource = 'L'
```

```python
#specify zipcode (along with latitude, longitue, and timezone).
#[zipcode, latitude, longitude, localtz] = ['08640', 40.0039, -74.6178, 'US/Eastern']
#[zipcode, latitude, longitude, localtz] = ['08641', 40.0449, -74.5892, 'US/Eastern']
#[zipcode, latitude, longitude, localtz] = ['92562', 33.5686, -117.2530, 'US/Pacific']
[zipcode, latitude, longitude, localtz] = ['92563', 33.5712, -117.1540, 'US/Pacific']
```

```python
#import required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
#from datetime import datetime, tzinfo, dat
import datetime
#from pytz import timezone
import pytz
from pytz import timezone
import pickle #for quick read/write list

#plotting in Notebook
%matplotlib inline
```

```python

dataDir = 'C:\Users\Admin\Dropbox\Active\EnergyProject\Thesis\PVreadingsStudies\main'
metadata = pd.read_csv(mainDir+ '/data/' + datasource +'/' + zipcode +'/metadata.csv',
                             dtype={'componentId':'object','zip':'object'})
#preserve raw metadata. We will modify it later
directory = mainDir + '/data/'+datasource+'/'+zipcode
metadata.to_csv(directory+"/metadata_original.csv",index=False)
timeseriesdata = pd.read_csv(mainDir+ '/data/' + datasource +'/'+ zipcode+'/timeseriesdata.csv',
                             dtype={'componentId':'object'})
```

```python
#minute adjustment to be 00,15,30 or 45#
timeseriesdata['tsUTC'] = timeseriesdata['tsUTC'].map(lambda x: x[0:14] + str((int(x[14:16])/15)*15).zfill(2) + ":00")
```

```python
#remove possible duplicated entries by averaging them
timeseriesdata = (timeseriesdata.groupby(['tsUTC','componentId'])).agg({"power":np.mean})
timeseriesdata.reset_index(inplace=True)
#pivot table
dat = timeseriesdata.pivot(index='tsUTC',columns='componentId',values='power')
dat.reset_index(inplace=True)

#Convert UTC time to local time. Keep time in string format for easy handling
dat['tsLocal'] = dat.tsUTC.map(lambda x: datetime.datetime.strptime(x,"%Y-%m-%d %H:%M:%S").replace(tzinfo=pytz.utc)\
                            .astimezone(pytz.timezone(localtz)).strftime("%Y-%m-%d %H:%M:%S"))
dat.drop('tsUTC', axis=1, inplace=True)
#Note that now tsUTC is a column, not an index and data frame.
```

Here one can look at power profile for each PV component ID in a given day. For example, 24st September 2014 (around September Equinox). Observe similarity and difference in power profiles from several components. 

```python
DateOfInterest = "2014-09-24"
df = dat[dat.tsLocal.map(lambda x: x[0:10] == DateOfInterest)]
df.tsLocal = df.tsLocal.map(lambda x: datetime.datetime.strptime(x,"%Y-%m-%d %H:%M:%S"))
df.set_index('tsLocal',inplace=True)
df.plot(legend=False)
plt.ylabel('Power [kw]')
```

## The first time stamp identification and data sorting

First, we sort data table by local time of reading and convert local time as an index.

```python
dat.sort(['tsLocal'], ascending = [1], inplace=True)
dat.set_index('tsLocal',inplace=True)
```

According to data frame, we identify the first time stamp for each component. We remove components that contain no power readings (so no first time stamp). Then we can sort columns according to this time stamp.

```python
[dat.shape[1],metadata.shape[0]]
```

```python
FirstTimestampList = []
j = 0
while j<dat.shape[1]:
    if np.sum(dat.iloc[:,j].notnull())>0: #check if the component has any meaningful reading at all
        FirstTimestampList.append(dat.index.tolist()[[i for i, elem in enumerate(dat.iloc[:,j].notnull()) if elem][0]])
        j = j+1
    else: #if not, remove that component out
        name_to_remove = dat.columns.values[j]
        dat.drop(name_to_remove, axis=1, inplace=True)
        metadata = metadata[metadata.componentId != name_to_remove]
ComponentList = list(dat.columns.values)
FirstTimestampTable = pd.DataFrame(0, index = np.arange(len(ComponentList)), columns = ['componentId','FirstTimestamp'])
FirstTimestampTable['componentId'] = ComponentList
FirstTimestampTable['FirstTimestamp'] = FirstTimestampList
FirstTimestampTable.sort(['FirstTimestamp'],ascending = [1], inplace=True)
dat = dat[list(FirstTimestampTable.componentId)]
```

```python
[dat.shape[1],metadata.shape[0]]
```

We can combine FirstTimestampTable with metadata and sort rows according to first time stamp. 

```python
if 'FirstTimestamp' not in list(metadata.columns.values):
    metadata = pd.merge(metadata,FirstTimestampTable,on=['componentId'])
    metadata.sort(['FirstTimestamp'],ascending = [1], inplace=True)
```

```python
#convert tsUTC back to column.
dat.reset_index(inplace=True)
```

## Scope and sun's altitude selection

Here we would like to data selection based on altitude of the sun. Here we coonstruct a data table with timestamp span matches our data set. This table should contain sun's altitude and azimuth angle. We may merge this table to our data table and select  only readings when the sun is up (altitude $>0$).

```python
#specify scope
StartScopeDate = '2013-01-01'
EndScopeDate = '2015-01-01'
StartScopeTime =  datetime.datetime.strptime(StartScopeDate,"%Y-%m-%d")
EndScopeTime =  datetime.datetime.strptime(EndScopeDate,"%Y-%m-%d")
```

```python
#Generate data frame from start to end, along with altitude and azimuth
tslist = []
nxt = StartScopeTime
while nxt < EndScopeTime:
    tslist.append(nxt)
    nxt += datetime.timedelta(minutes=15)
    
Scope = pd.DataFrame(0, index = np.arange(len(tslist)), columns = ['tsLocal','altitude','azimuth'])
Scope['tsLocal'] = tslist

#need to install Pysolar in command prompt 'pip install Pysolar' first
import Pysolar
Scope['altitude'] = Scope.tsLocal.map(lambda x: Pysolar.GetAltitude(latitude, longitude, 
                                                                    timezone(localtz).localize(x).astimezone(pytz.utc)))

#Consider only when the sun's altitude is high.
Scope = Scope[Scope.altitude>0]
Scope['azimuth'] = Scope.tsLocal.map(lambda x: Pysolar.GetAzimuth(latitude, longitude, 
                                                                    timezone(localtz).localize(x).astimezone(pytz.utc)))
#covert to string to easy handling
Scope['tsLocal'] = Scope.tsLocal.map(lambda x: x.strftime("%Y-%m-%d %H:%M:%S")) 
```

```python
#Now we can merge using Scope Table as a backbone of data structure.
data = pd.merge(Scope,dat,on=['tsLocal'], how = 'left') 
```

Now $\texttt{data}$ contains power readings. The rows represent all time stamps in 15 minutes while sun is up for all days of interest. The columns (except for time stamp, azimuth, and altitudes) represent all functioning components in the system. $\texttt{metadata}$ contains same information with first time stamp in addition.
