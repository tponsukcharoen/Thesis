# 15-min rolling prediction heurestics

According to analysis above, one can predict $s_i$ once $s_{i-1}$ and $s_{i-2}$ are known based on regression coefficient that we found earlier. It is not clear how to proceed in 15-minute steps. Here is a heurestic for each day operation:

1. Predict $s_0$ to be a constant, say the mean of $s_0$. Predict $s_1$ using predicted $s_2$, and so on.
2. Check where the 15-min reading lies, then use interpolation of predicted values on the grid to predict the measurement.
3. Once the true value is realized, use interpolation of measurement to update the values on the grid. 
4. Any time the $s_i$ is updated, update subsequent $s_i$ according to the regression.

To illustrate how it works. Let's use p-ratios before the last quarter of 2014 to train regression and do 15-min rolling prediction on the last quarter of 2014.

```python
#select data for training
StartDate = datetime.date(2014,10,1)
X = ((PratioRef[PratioRef.Date<StartDate].dropna()).set_index(['Date']))

DateList = list(dsummary.Date[dsummary.Date<StartDate])
y0list = []
for DateOfInterest in DateList:
    df = data[data.tsLocal.map(lambda x: x.date())==DateOfInterest]
    y0list.append(df.iloc[0].pratio) 
```

```python
#fixed value for s_0
y0 = np.nanmean(y0list)
y0low = np.nanpercentile(y0list,2.5) 
y0high = np.nanpercentile(y0list,97.5) 
y0, y0low, y0high
```

```python
ModelList = [0] * 48

previous_values = 2
XX = X.iloc[:,range(1)]
y = X.iloc[:,1]
ModelList[0] = sklearn.linear_model.LinearRegression().fit(XX,y)

for i in range(2,49):
    y = X.iloc[:,i]
    XX = X.iloc[:,range(i-previous_values,i)]
    ModelList[i-1] = sklearn.linear_model.LinearRegression().fit(XX,y)
    
#fixed value for s_0
y1 = np.mean(X.iloc[:,0])
y1low = np.percentile(X.iloc[:,0],2.5) 
y1high = np.percentile(X.iloc[:,0],97.5) 
y1, y1low, y1high
```

```python
#define function for rolling regression
def roll_regression(p,location): #location counted from 0,...,50 [-1,1]
    for i in range(location,49):
        if i==2:
            p[i] = ModelList[i-2].predict(p[0])
        else:
            p[i] = ModelList[i-2].predict(p[range(i-2,i)])
    return p;
    
```

```python
#set up directory for plot
directory = outputDir + '/15minrolling/'
if not os.path.exists(directory):
    os.makedirs(directory)

DateList = list(dsummary.Date[dsummary.Date>=StartDate])
for DateOfInterest in [DateList[0]]:
    #StartDate = datetime.date(2014,10,1)
    #X = ((PratioRef[PratioRef.Date<StartDate].dropna()).set_index(['Date']))
    #DateOfInterest = datetime.date(2014,10,1)
    df = data[data.tsLocal.map(lambda x: x.date())==DateOfInterest]
    df['pratio_predict']=float('NaN')

    pratiolist = np.array([y0] + [y1] + [1]*48 + [1])

    pratiolist = roll_regression(pratiolist,2)
    s_old = -1
    pratio_old = 1

    plt.clf()
    fig = plt.figure(figsize=(12, 6))
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
    ax1.plot(df.s,df.pratio,'.',color='green')
    ax2.plot(df.s,df.power,'.',color='green')
    ax2.plot(df.s,df.mpower,color='black')
    #ax1.set_ylim(0,1.1)
    ax1.set_xlim(-1.1,-0.5)
    ax1.set_title('pratio')
    ax2.set_ylim(0,1.1)
    ax1.axvline(x=-1.0+2.0/50*1)
    ax1.axvline(x=-1.0+2.0/50*2)
    ax2.axvline(x=-1.0+2.0/50*1)
    ax2.axvline(x=-1.0+2.0/50*2)

    for j in range(0,5):#df.shape[0]):
        #show current pratiolist
        ax1.plot(map(lambda x : -1.0+2.0/50*x, range(0,51)),pratiolist,color='blue', alpha = 0.1)
        #prediction phase
        s = df.iloc[j].s
        i1 = math.floor((s+1.0)*50/2.0)
        i2 = i1+1
        s1 = -1.0+2.0/50*i1
        s2 = -1.0+2.0/50*i2
        pratio1 = pratiolist[i1]
        pratio2 = pratiolist[i2]
        df.pratio_predict.iloc[j] = pratio1 + (pratio2-pratio1)/(s2-s1)*(s-s1)
        #plot predicted value
        ax1.plot(s,df.pratio_predict.iloc[j],'.',color='red')
        ax2.plot(s,df.pratio_predict.iloc[j]*(df.mpower.iloc[j]+shift)-shift,'.',color='red')

        #Update known value, interpolate, and project to future s in the grid..
        pratio = df.iloc[j].pratio
        #check which is sgrid is the range of interpolation
        bool_list = map(lambda x : (-1.0+2.0/50*x >= s_old)&(-1.0+2.0/50*x < s), range(0,51))
        indexlist = [i for i, elem in enumerate(bool_list) if elem]
        if len(indexlist)>0:
            for i in indexlist:
                pratiolist[i] = pratio_old + (pratio-pratio_old)/(s-s_old)*(-1.0 + 2.0/50*i-s_old)
            #update the projection
            pratiolist = roll_regression(pratiolist,max(indexlist)+1)
        else:
            #back updating
            pratiolist[i1] = pratio_old + (pratio-pratio_old)/(s-s_old)*(-1.0 + 2.0/50*i1-s_old)
            #forward updating (needed in case we don't have model to regress.)
            pratiolist[i2] = pratio_old + (pratio-pratio_old)/(s-s_old)*(-1.0 + 2.0/50*i2-s_old)
            pratiolist = roll_regression(pratiolist,int(i2))
        s_old = s
        pratio_old = pratio

    ax2.plot(df.s,np.absolute(df.pratio_predict*(df.mpower+shift)-shift-df.power),color='blue')
    ax2.set_title('power: MAE='+ str(np.mean(np.absolute(df.power-(df.pratio_predict*(df.mpower+0.1)-0.1)))) \
                  +', mean Power=' + str(np.mean(df.power)))
    
    #filename = DateOfInterest.strftime("%Y-%m-%d") +'.png'
    #fig.savefig(directory+filename)
```
