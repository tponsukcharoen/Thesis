Here we define power ratio for each power reading. We add 0.01 to avoid infinity.

```python
data['pratio'] = (data.power+0.01)/(data.mpower+0.01)
```

```python
data[:1]
```

Now we would like to construct consistent measure of pratio across different values of s. Here we use 49 values of s between -1 and 1 as a reference and interpolate power ratio to these values. 

```python
PratioRef = pd.DataFrame(columns = ['Date'] + map(lambda x: str(x),range(1,50)))
PratioRef['Date']=dsummary.Date
```

```python
for i in range(0, dsummary.shape[0]): 
    DateOfInterest = dsummary.Date.iloc[i]
    E = dsummary.E.iloc[i]
    if (not math.isnan(E)):
        df = data[data.tsLocal.map(lambda x: x.date()) == DateOfInterest]
        for j in range(1,50):
            starget = -1.0 + 2.0/50*j
            s1 = max(list(df.s[df.s <= starget]))
            s2 = min(list(df.s[df.s >= starget]))
            i1 = list(df.s).index(s1)
            i2 = list(df.s).index(s2)
            pratio1 = list(df.pratio)[i1]
            pratio2 = list(df.pratio)[i2]
            #avoid using values at end point (sun rise/ sun set). 
            if i1==0:
                ptarget = pratio2
            elif i2==len(list(df.s))-1:
                ptarget = pratio1
            else:
                ptarget = pratio1 + (pratio2-pratio1)/(s2-s1)*(starget-s1)
            
            PratioRef[str(j)].iloc[i] = ptarget
```

# Basic regression to predict next pratio

One basic problem we would like to solve in daily operation is that, given pratio throughout the day, what is the next pratio? One may use pratio values for all previous s's as an input in regression. Let's use linear regression so that we can see the influence of different previous values.

```python
CoefTable = pd.DataFrame([[float('NaN')]*48 for i in range(48)],columns = map(lambda x: str(x),range(1,49)))
RsquaredTable = pd.DataFrame([float('NaN')]*48,columns = ['Rsquared'])
```

```python
model = sklearn.linear_model.LinearRegression()

for i in range(1,49):
    y = X.iloc[:,i]
    XX = X.iloc[:,range(0,i)]
    mod = model.fit(XX,y)
    CoefTable.iloc[i-1,range(0,i)] = mod.coef_
    RsquaredTable.iloc[i-1] = model.score(XX,y)
```

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(121)
ax.set_title('regression coefficient')
plt.imshow(CoefTable)
ax.set_aspect('equal')
cbar = plt.colorbar()
cbar.ax.set_ylabel('coef (-)')
ax = fig.add_subplot(122)
ax.plot(RsquaredTable)
```

```python
model = sklearn.linear_model.Lasso(alpha=0.0001) #0.00005,0.0001

for i in range(1,49):
    y = X.iloc[:,i]
    XX = X.iloc[:,range(0,i)]
    mod = model.fit(XX,y)
    CoefTable.iloc[i-1,range(0,i)] = mod.coef_
    RsquaredTable.iloc[i-1] = model.score(XX,y)
    
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(121)
ax.set_title('regression coefficient')
plt.imshow(CoefTable)
ax.set_aspect('equal')
cbar = plt.colorbar()
cbar.ax.set_ylabel('coef (-)')
ax = fig.add_subplot(122)
ax.plot(RsquaredTable)
```
