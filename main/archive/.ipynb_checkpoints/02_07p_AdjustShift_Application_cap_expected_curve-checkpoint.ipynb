We can use this to smoothen expected performance ratio and give a good expected power curve depending on time of day (via s) and day of year (via DBW). This requires 2 dimensional interpolation with periodic domain. One simple way is to use a triangular mesh.

c2 oo oo oo c1

oo oo oo oo oo

oo oo XY oo oo

oo oo oo oo oo

c3 oo oo oo c4


```python
def find_expectedr(row):
    #if (math.isnan(row['s']))|(row['DBW'].isnan()):
    #    return  float('NaN')
    #else:
    s = row['s']
    DBW = row['DBW']
    x = (s+1)/2.*6
    y = (DBW+365)/365.*6
    X = round(x)
    Y = round(y)
    dX = x-X
    dY = y-Y
    Xfront = int(X*(not X==6) +0*(X==6))
    Xback = int((X-1)*(not X==0) +5*(X==0))
    Yfront = int(Y*(not Y==6) +0*(Y==6))
    Yback = int((Y-1)*(not Y==0) +5*(Y==0))

    [c1,c2,c3,c4] = [Matrix[Xfront][Yfront],Matrix[Xback][Yfront],Matrix[Xback][Yback],Matrix[Xfront][Yback]]

    value = 0.5*(c1 + c3)
    if dY > dX:
        value = value + (dY)*(c2-c3) + (dX)*(c1-c2)
    else:
        value = value + (dY)*(c1-c4) + (dX)*(c4-c3)
    return value;


#data['expectedr'] = data.apply(find_expectedr, axis=1)
d.loc[:,'expectedr'] = float('NaN')
bool_list = (d['s'].notnull())&(d['DBW'].notnull())
d.loc[bool_list,'expectedr'] = d[bool_list].apply(find_expectedr, axis=1)
```

```python
#For thesis, examples of maximum daily power curve

DateOfInterest = "2014-06-24" #21 #24 example for good days
df = d[d.tsLocal.map(lambda x: x.date()) == (datetime.datetime.strptime(DateOfInterest,"%Y-%m-%d")).date()]

fig = plt.figure(figsize=(10,6))
ax1  = fig.add_subplot(111)

ax1.plot(df.tsLocal.map(lambda x: x.time()),df.power)
ax1.plot(df.tsLocal.map(lambda x: x.time()),df.mpower,color='red')
ax1.plot(df.tsLocal.map(lambda x: x.time()),df.expectedr*df.mpower,color='green')
ax1.set_xlabel('Time of Day')
ax1.set_ylabel('Normalized power [-]')
ax1.set_ylim([0,1.2])
ax1.set_xlim([datetime.datetime.strptime(DateOfInterest +" 05:00:00","%Y-%m-%d %H:%M:%S").time(),\
              datetime.datetime.strptime(DateOfInterest +" 21:00:00","%Y-%m-%d %H:%M:%S").time()])
#plt.title('Average normalized power in zipcode ' + zipcode +'\n' +str(DateOfInterest))
#plt.show()
```

These green curves are useful in day-ahead or long term planning.
