Here we define power ratio for each power reading. We add 0.1 to avoid infinity.

```python
shift = 0.01
data['pratio'] = (data.power+shift)/(data.mpower+shift)
```

```python
data[:1]
```

Now we would like to construct consistent measure of pratio across different values of s. Here we use 51 values of s from -1 to 1 as a reference and interpolate power ratio to these values. 

```python
PratioRef = pd.DataFrame(columns = ['Date'] + map(lambda x: str(x),range(0,51)))
PratioRef['Date']=dsummary.Date
```

```python
for i in range(0, dsummary.shape[0]): 
    DateOfInterest = dsummary.Date.iloc[i]
    E = dsummary.E.iloc[i]
    if (not math.isnan(E)):
        df = data[data.tsLocal.map(lambda x: x.date()) == DateOfInterest]
        PratioRef['0'].iloc[i] = df.pratio.iloc[0]
        PratioRef['50'].iloc[i] = df.pratio.iloc[-1]        
        for j in range(1,50):
            starget = -1.0 + 2.0/50*j
            s1 = max(list(df.s[df.s <= starget]))
            s2 = min(list(df.s[df.s >= starget]))
            i1 = list(df.s).index(s1)
            i2 = list(df.s).index(s2)
            pratio1 = list(df.pratio)[i1]
            pratio2 = list(df.pratio)[i2]
            
            #interpolation gives a false impression of correlation
            #if s1==s2:
            #    ptarget = pratio1
            #else:
            #    ptarget = pratio1 + (pratio2-pratio1)/(s2-s1)*(starget-s1)
            
            #rounding gives a false impression of correlation when the values are used multiple times (in the middle part)
            if np.absolute(s1-starget) <= np.absolute(s2-starget):
                ptarget = pratio1
            else:
                ptarget = pratio2
                
            PratioRef[str(j)].iloc[i] = ptarget
```

## Correlation

Correlation between values on the grid on the same days is deceptive since it comes from interpolation or redundant use of raw data especially the middle of the grid. Hence, we should not infer any correlation from values on the grid on the same day. The correlation between days is still somewhat applicable.

```python
c_s = PratioRef.drop(['Date'],axis=1).astype(float).corr(method='pearson', min_periods=1)
c_days = PratioRef.drop(['Date'],axis=1).astype(float).transpose().corr(method='pearson', min_periods=1)
```

# Basic regression to predict next p-ratio: This is all deceptive due to interpolation/ redundant use of raw data

One basic problem we would like to solve in daily operation is that, given pratio throughout the day, what is the next pratio? One may use pratio values for all previous s's as an input in regression. Let's use linear regression so that we can see the influence of different previous values.

```python
#X = ((PratioRef.fillna(0)).set_index(['Date']))
X = ((PratioRef.dropna()).set_index(['Date']))
```

```python
CoefTable = pd.DataFrame([[float('NaN')]*51 for i in range(50)],columns = map(lambda x: str(x),range(0,51)))
RsquaredTable = pd.DataFrame([float('NaN')]*50,columns = ['Rsquared'])
InterceptTable = pd.DataFrame([float('NaN')]*50,columns = ['Intercept'])

model = sklearn.linear_model.LinearRegression()

for i in range(1,51):
    y = X.iloc[:,i]
    XX = X.iloc[:,range(0,i)]
    mod = model.fit(XX,y)
    CoefTable.iloc[i-1,range(0,i)] = mod.coef_
    InterceptTable.iloc[i-1]= model.intercept_
    RsquaredTable.iloc[i-1] = model.score(XX,y)
    
fig = plt.figure(figsize=(18, 6))
ax = fig.add_subplot(131)
ax.set_title('regression coefficient')
plt.imshow(CoefTable)
ax.set_aspect('equal')
cbar = plt.colorbar()
cbar.ax.set_ylabel('coef (-)')
ax = fig.add_subplot(132)
ax.plot(range(1,51),RsquaredTable,'.')
ax.set_title('Rsq')
ax = fig.add_subplot(133)
c = CoefTable
i=0
l = np.diagonal(c, offset=-i, axis1=0, axis2=1)
ax.plot(range(i,len(l)+i),l,color='red')
i=1
l = np.diagonal(c, offset=-i, axis1=0, axis2=1)
ax.plot(range(i,len(l)+i),l,color='green')
i=2
l = np.diagonal(c, offset=-i, axis1=0, axis2=1)
ax.plot(range(i,len(l)+i),l,color='blue')
i=3
l = np.diagonal(c, offset=-i, axis1=0, axis2=1)
ax.plot(range(i,len(l)+i),l,color='purple')
ax.set_title('Coef')
ax.plot(InterceptTable,color='black')
```

From the analysis above, one can see that we should predict values at $s=-1$ and $s=1$ separately. In addition, values at $s=-1$ should be excluded from the regressions.

Let's use only p-values from previous 2 s-values as predictors except what is described above. 

```python
CoefTable = pd.DataFrame([[float('NaN')]*51 for i in range(50)],columns = map(lambda x: str(x),range(0,51)))
RsquaredTable = pd.DataFrame([float('NaN')]*51,columns = ['Rsquared'])
InterceptTable = pd.DataFrame([float('NaN')]*51,columns = ['Intercept'])
model = sklearn.linear_model.LinearRegression()
previous_values = 2
for i in range(2,50): #we generate regression for index 2 to 49 only.
    y = X.iloc[:,i]
    XX = X.iloc[:,range(max(1,i-previous_values),i)]
    mod = model.fit(XX,y)
    CoefTable.iloc[i,range(max(1,i-previous_values),i)] = mod.coef_
    InterceptTable.iloc[i]= model.intercept_
    RsquaredTable.iloc[i] = model.score(XX,y)
    
fig = plt.figure(figsize=(18, 6))
ax = fig.add_subplot(131)
ax.set_title('regression coefficient')
plt.imshow(CoefTable)
ax.set_aspect('equal')
cbar = plt.colorbar()
cbar.ax.set_ylabel('coef (-)')
ax = fig.add_subplot(132)
ax.plot(range(0,51),RsquaredTable,'.')
ax.set_title('Rsq')
ax = fig.add_subplot(133)
c = CoefTable
i=1
l = np.diagonal(c, offset=-i, axis1=0, axis2=1)
ax.plot(range(i,len(l)+i),l,color='red')
i=2
l = np.diagonal(c, offset=-i, axis1=0, axis2=1)
ax.plot(range(i,len(l)+i),l,color='green')
ax.plot(InterceptTable,color='black')
    
```
