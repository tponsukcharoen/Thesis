Measure daily ratio shift

```python
#select data to analyze
StartDate = datetime.date(2014,1,1)
TestData = d[d.tsLocal.map(lambda x: x.date()) >=StartDate].set_index('tsLocal')
TestDateList = list(set(TestData.index.map(lambda x: x.date())))
#set up date list and summary table
TestDateList.sort()

RatioTable = pd.DataFrame(float('NaN'), index = np.arange(len(TestDateList)), \
                        columns = ['day_mean','day_d','week_mean','week_d','year_mean','year_d','blend_mean',\
                                   'blend_d','peak','dayload','weekday'])
RatioTable['Date']=TestDateList
```

```python
#Loop for ratio calculation
for i in range(0,RatioTable.shape[0]):
    DailyData = TestData[TestData.index.map(lambda x: x.date()) == RatioTable.Date[i]]
    RatioTable.loc[i,'weekday'] = DailyData.weekday.iloc[0]
    d_series = DailyData['load'].dropna()
    if len(d_series)>0:
        RatioTable.loc[i,'dayload'] = np.sum(d_series)
        RatioTable.loc[i,'peak'] = np.max(d_series)
        
    d_series = (DailyData['load']/DailyData['load-1d']).dropna()
    if len(d_series)>0:
        RatioTable.loc[i,'day_mean'] = np.mean(d_series)
        RatioTable.loc[i,'day_d'] = np.max(np.absolute(d_series-np.mean(d_series)))
        
    d_series = (DailyData['load']/DailyData['load-1w']).dropna()
    if len(d_series)>0:
        RatioTable.loc[i,'week_mean'] = np.mean(d_series)
        RatioTable.loc[i,'week_d'] = np.max(np.absolute(d_series-np.mean(d_series)))
    
    d_series = (DailyData['load']/DailyData['load-1y']).dropna()
    if len(d_series)>0:
        RatioTable.loc[i,'year_mean'] = np.mean(d_series)
        RatioTable.loc[i,'year_d'] = np.max(np.absolute(d_series-np.mean(d_series)))
    
    
    if DailyData.weekday.iloc[0] == 0: #Monday
        l = (DailyData['load-1w'])
        #l = (DailyData['load-3d']+DailyData['load-4d']+DailyData['load-5d']+DailyData['load-6d']+DailyData['load-1w'])/5
        l = (l+DailyData['load-1y'])/2
    elif DailyData.weekday.iloc[0] == 1: #Tuesday
        l = (DailyData['load-1d'])
        #l = (DailyData['load-1d']+DailyData['load-4d']+DailyData['load-5d']+DailyData['load-6d']+DailyData['load-1w'])/5
        #l = (l+DailyData['load-1y'])/2
    elif DailyData.weekday.iloc[0] == 2:
        l = (DailyData['load-1d'])
        #l = (DailyData['load-1d']+DailyData['load-2d']+DailyData['load-5d']+DailyData['load-6d']+DailyData['load-1w'])/5
        #l = (l+DailyData['load-1y'])/2
    elif DailyData.weekday.iloc[0] == 3:
        l = (DailyData['load-1d'])
        #l = (DailyData['load-1d']+DailyData['load-2d']+DailyData['load-3d']+DailyData['load-6d']+DailyData['load-1w'])/5
        #l = (l+DailyData['load-1y'])/2
    elif DailyData.weekday.iloc[0] == 4: 
        l = (DailyData['load-1d'])
        #l = (DailyData['load-1d']+DailyData['load-2d']+DailyData['load-3d']+DailyData['load-4d']+DailyData['load-1w'])/5
        #l = (l+DailyData['load-1y'])/2
    elif DailyData.weekday.iloc[0] == 5: #Saturday
        l = (DailyData['load-1w'])
        #l = (DailyData['load-6d']+DailyData['load-1w'])/2
        l = (l+DailyData['load-1y'])/2
    else:
        l = (DailyData['load-1d'])
        #l = (DailyData['load-1d']+DailyData['load-1w'])/2
        #l = (l+DailyData['load-1y'])/2
    
    #blend = 0
    
    #l = DailyData['load-1d']  
    d_series = (DailyData['load']/l).dropna()
    
    #if (DailyData.weekday.iloc[0] == 0)|(DailyData.weekday.iloc[0] == 5):
        #d_series = DailyData['load']/l
        #d_series = (DailyData['load']/(blend*l+(1-blend)*DailyData['load-1w'])).dropna()
        #d_series = (DailyData['load']/(0.1*DailyData['load-1d']+0.9*DailyData['load-1w'])).dropna()
    #else:
        #d_series = l
        #d_series = (DailyData['load']/((1-blend)*l+blend*DailyData['load-1w'])).dropna()
        #d_series = (DailyData['load']/(0.9*DailyData['load-1d']+0.1*DailyData['load-1w'])).dropna()
        
    if len(d_series)>0:
        RatioTable.loc[i,'blend_mean'] = np.mean(d_series)
        RatioTable.loc[i,'blend_d'] = np.max(np.absolute(d_series-np.mean(d_series)))
```

## Model I: Naive model: current value equals to previous hour value

```python
def scanbyweekday(d,i):
    d_cal = d[['load','load_predict']][d.weekday==i].dropna()
    return [np.mean(np.absolute((d_cal['load']-d_cal['load_predict']))/d_cal['load']),\
        max(np.absolute((d_cal['load']-d_cal['load_predict'])/d_cal['load']))];
```

```python
#actual value
d['load_predict']=float('NaN')
d['load_predict']=d['load-1h']
d_cal = d[['load','load_predict']].dropna()
[np.mean(np.absolute((d_cal['load']-d_cal['load_predict']))/d_cal['load']),\
 max(np.absolute((d_cal['load']-d_cal['load_predict'])/d_cal['load']))]
```

```python
#ratio of previous day
d['load_predict']=float('NaN')
d['load_predict']=d['load-1h']/d['load-1d-1h']*d['load-1d']
d_cal = d[['tsLocal','load','load_predict']].dropna()
[np.mean(np.absolute((d_cal['load']-d_cal['load_predict']))/d_cal['load']),\
 max(np.absolute((d_cal['load']-d_cal['load_predict'])/d_cal['load']))]
```

```python
#Daily Average error
RatioTable['AverageDailyError']=float('NaN')
#Loop for ratio calculation
for i in range(0,RatioTable.shape[0]):
    DailyData = d_cal[d_cal.tsLocal.map(lambda x: x.date()) == RatioTable.Date[i]]
    DailyData['percent_error'] = np.absolute(DailyData['load']-DailyData['load_predict'])/DailyData['load']
    d_series = DailyData['percent_error'].dropna()
    if len(d_series)>0:
        RatioTable.loc[i,'percent_error'] = np.mean(d_series)
        
#plt.plot(VariabilityTable['E'],VariabilityTable['V'],'.')
fig = plt.figure(figsize=(16, 8))
ax1 = fig.add_subplot(211)
plt.plot(RatioTable.index,RatioTable['percent_error'],'.')
#d_series.plot(ylim=[0,1],style='.')
plt.vlines(x = RatioTable.index, ymin= 0, ymax = RatioTable['percent_error'])
#plt.ylabel('peak load')
```
